```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# The path to our CSV file
file = "data/most-recent-cohorts-all-data-elements-csv-5.csv"

# Read our Kickstarter data into pandas
df5 = pd.read_csv(file)
df5.head()
```

```python
file2="data/most-recent-cohorts-treasury-elements-csv-3.csv"
df3=pd.read_csv(file2)
df3
```

### Data Transformation - take select columns needed

```python
clean_df5=df5[["UNITID","OPEID","INSTNM","HIGHDEG","CONTROL","SCH_DEG","UGDS","PPTUG_EF","ADM_RATE","CITY","STABBR",
              "ZIP","LATITUDE","LONGITUDE","INSTURL","POVERTY_RATE","UNEMP_RATE","FAMINC","MD_FAMINC","FAMINC_IND",
              "TUITFTE","INEXPFTE","AVGFACSAL","NPT4_PUB","NPT4_PRIV","COSTT4_A","COSTT4_P","RELAFFIL","ACCREDAGENCY"]]
clean_df5
```

```python
clean_df3=df3[["UNITID","OPEID","PCT_WHITE","PCT_BLACK","PCT_ASIAN","PCT_HISPANIC","PCT_BA","PCT_BORN_US"]]
clean_df3
```

```python
# Merge all data frames into a master data frame to work with
master_df = clean_df5.merge(clean_df3, on=['UNITID', 'OPEID'])
master_df
```

### Create Seperate Table Specific CSVs

```python
profiles_df = master_df[['UNITID', 'OPEID', 'INSTNM', 'CONTROL', 'SCH_DEG', 'RELAFFIL', 'HIGHDEG', 'INSTURL']]
profiles_df.head()
```

```python
costs_df = master_df[['UNITID','TUITFTE', 'INEXPFTE', 'AVGFACSAL', 'NPT4_PRIV', 'NPT4_PUB', 'COSTT4_A', 'COSTT4_P']]
costs_df.head()
```

```python
demographics_df = master_df[['UNITID','PCT_WHITE', 'PCT_BLACK', 'PCT_ASIAN', 'PCT_HISPANIC', 'PCT_BA', 'PCT_BORN_US']]
demographics_df.head()
```

```python
enrollment_df = master_df[['UNITID','ADM_RATE', 'UGDS', 'PPTUG_EF']]
enrollment_df.head()
```

```python
geography_df = master_df[['UNITID','CITY', 'STABBR', 'ZIP', 'LATITUDE', 'LONGITUDE']]
geography_df.head()
```

```python
need_df = master_df[['UNITID','FAMINC', 'MD_FAMINC', 'FAMINC_IND']]
need_df.head()
```

```python
outcomes_df = master_df[['UNITID','POVERTY_RATE', 'UNEMP_RATE']]
outcomes_df.head()
```

### Connect to local database


```python
rds_connection_string = "postgres:postgres@localhost:5432/etl_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Upload Data


```python
profiles_df.to_sql(name='Profiles', con=engine, index=False)
```

```python
costs_df.to_sql(name='Costs', con=engine, index=False)
```

```python
demographics_df.to_sql(name='Demographics', con=engine, index=False)
```

```python
enrollment_df.to_sql(name='Enrollment', con=engine, index=False)
```

```python
geography_df.to_sql(name='Geography', con=engine, index=False)
```

```python
need_df.to_sql(name='Need', con=engine, index=False)
```

```python
outcomes_df.to_sql(name='Outcomes', con=engine, index=False)
```
